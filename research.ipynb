```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
load_dotenv()

llm = ChatGroq(
    api_key = os.getenv("GROQ_API_KEY"),
    model_name = "llama-3.1-8b-instant",
    temperature = 0.1)

prompt = """
How will you evaluate a mathematical problem?

Lets say:- 3 x 5 or integration of logx with respect to x
How will you verify whether the answer is correct?
"""

response = llm.invoke(prompt).content
```
